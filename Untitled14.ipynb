```python
import pandas as pd
import matplotlib.pyplot as plt

df['price_per_m2'] = df['price'] / df['area_usable']

station_stats = df.groupby('nearest_metro').agg({
    'price': 'mean',
    'price_per_m2': 'mean'
})

station_stats_sorted = station_stats.sort_values(by='price_per_m2', ascending=False)

print("TOP 10 MOST EXPENSIVE STATIONS:")
print(station_stats_sorted.head(10))

print("\nTOP 10 CHEAPEST STATIONS:")
print(station_stats_sorted.tail(10))

plt.figure(figsize=(14, 6))
plt.bar(station_stats_sorted.index[:10], station_stats_sorted['price_per_m2'][:10], color='red')
plt.title('Top 10 Most Expensive Metro Stations (Average price per m2)')
plt.ylabel('Price (CZK)')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(14, 6))
plt.bar(station_stats_sorted.index[-10:], station_stats_sorted['price_per_m2'][-10:], color='green')
plt.title('Top 10 Cheapest Metro Stations (Average price per m2)')
plt.ylabel('Price (CZK)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df['price_per_m2'] = df['price'] / df['area_usable']

line_comparison = df.groupby('nearest_metro_line')['price_per_m2'].mean().sort_values(ascending=False)

print("Average price per m2 by metro line:")
print(line_comparison)

plt.figure(figsize=(10, 6))
line_comparison.plot(kind='bar', color=['green', 'yellow', 'red'])
plt.title('Average Apartment Price per Metro Line')
plt.xlabel('Metro Line')
plt.ylabel('Average Price per m2 (CZK)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df['price_per_m2'] = df['price'] / df['area_usable']

metro_index = df.groupby('nearest_metro').agg({
    'price_per_m2': 'mean',
    'price': 'count'
}).rename(columns={'price': 'count'})

metro_index = metro_index[metro_index['count'] > 5]

metro_index = metro_index.sort_values(by='price_per_m2', ascending=False)

print("Metro Index - Top 10 Locations by Price per m2:")
print(metro_index.head(10))

plt.figure(figsize=(12, 8))
plt.barh(metro_index.index[:15], metro_index['price_per_m2'][:15], color='skyblue')
plt.gca().invert_yaxis()
plt.title('Metro Index: Top 15 Most Expensive Locations')
plt.xlabel('Average Price per m2 (CZK)')
plt.show()
```
